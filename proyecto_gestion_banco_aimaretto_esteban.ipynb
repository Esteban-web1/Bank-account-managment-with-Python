## 4Ô∏è‚É£  Resumen de la base de datos

**Problema.** El banco quiere conocer datos resumen de sus clientes. En particular le interesa saber:

* Cu√°ntos clientes hay en total.
* Si existen clientes con nombre repetidos y en tal caso, cu√°ntos.
* Cu√°les son los clientes m√°s j√≥venes.
* Cu√°les son los clientes que m√°s y menos saldo tienen.
* Cu√°l es la edad media de sus clientes.
* Cu√°l es el saldo medio de sus clientes.


Ejercicio 4.1: Crear una funci√≥n llamada `calcular_metricas` que tome de entrada el DataFrame `clientes_actualizado` y muestre en pantalla los resultados de los √≠tems solicitados. Nuevamente, agregar documentaci√≥n a esta funci√≥n


<br>

---

<br>

````python
#Resuelvo el Ejercicio 4.1:
def calcular_metricas(df):
    """
    Calcula y muestra en pantalla los resultados de los √≠tems solicitados.
    """
    # Cu√°ntos clientes hay en total.
    total_clientes = len(df)
    print(f"Total de clientes: {total_clientes}")
    # Si existen clientes con nombres repetidos, y en tal caso, cu√°ntos
    nombres_repetidos = df[df.duplicated(subset='nombre', keep=False)]
    if not nombres_repetidos.empty:
        print("Clientes con nombres repetidos:")
        print(nombres_repetidos)
    #Cu√°les son los clientes m√°s j√≥venes
    clientes_mas_jovenes = df[df['edad'] == df['edad'].min()]
    print("Clientes m√°s j√≥venes:")
    print(clientes_mas_jovenes)
    #Cu√°les son los clientes que m√°s y menos saldo tienen
    cliente_mas_saldo = df[df['saldo'] == df['saldo'].max()]
    cliente_menos_saldo = df[df['saldo'] == df['saldo'].min()]
    print("Cliente con m√°s saldo:")
    print(cliente_mas_saldo)
    print("Cliente con menos saldo:")
    print(cliente_menos_saldo)
    #Cu√°l es la edad media de sus clientes
    edad_media = df['edad'].mean()
    print(f"Edad media de los clientes: {edad_media:.2f}")
    #Cu√°l es el saldo medio de sus clientes
    saldo_medio = df['saldo'].mean()
    print(f"Saldo medio de los clientes: {saldo_medio:.2f}")
#compruebo que funcione el c√≥digo con el df de clientes_actualizado
calcular_metricas(clientes_actualizado)
````

Ejercicio 5.1: Crear un gr√°fico de barra que muestre los saldos de sus clientes.

Ejercicio 5.2: Crear un gr√°fico de barra que muestre la edad de sus clientes.

Ejercicio 5.3: Crear n gr√°fico de dispersi√≥n (scatter plot) que muestre la relaci√≥n entre la edad y el saldo de los clientes, lo que podr√≠a ayudar a entender si hay tendencias seg√∫n grupos etarios.

- Nombrar los ejes si corresponde (por ejemplo, "Nombre del cliente" en el eje X y "Saldo" en el eje Y para un gr√°fico de barras).

````python
# 5, primero importo la librer√≠a de matplot:
import matplotlib.pyplot as plt
````

````python
#Resuelvo el Ejercicio 5.1,creando una figura de barras que muestre el saldo de los clientes:
plt.figure(figsize=(10, 6))
#Cambio los colores de las barras seg√∫n superen 20000 en la variable saldo o no:
colores = ['green' if saldo > 20000 else 'red' for saldo in clientes_actualizado['saldo']]
plt.bar(clientes_actualizado['nombre'], clientes_actualizado['saldo'], color=colores)
#Nombre del t√≠tulo del gr√°fico
plt.title('Saldos de los clientes del banco')
#Nombre de las variables de los ejes
plt.xlabel('Nombre del cliente')
plt.ylabel('Saldo')
#Roto la inclinaci√≥n de los nombres de clientes para que no se superpongan
plt.xticks(rotation=90)
plt.show()
````

````python
#Resuelvo el Ejercicio 5.2, creando una figura por edades:
plt.figure(figsize=(10, 6))
#Cambio los colores de las barras que tengan clientes con edad mayor a 45:
colores = ['green' if edad > 45 else 'red' for edad in clientes_actualizado['edad']]
plt.bar(clientes_actualizado['nombre'], clientes_actualizado['edad'], color=colores)
#Nombro el t√≠tulo del gr√°fico
plt.title('Edades de los clientes del banco')
#Nombro las variables que van en los ejes
plt.xlabel('Nombre del cliente')
plt.ylabel('Edad')
#Roto la inclinaci√≥n de los nombres de clientes para que no se superpongan
plt.xticks(rotation=90)
plt.show()
````

````python
#Resuelvo el ejercicio 5.3, creando un gr√°fico de dispersi√≥n:
#Cambio el color de la observaci√≥n con mayor saldo:
colores = ['green' if saldo == clientes_actualizado['saldo'].max() else 'red' for saldo in clientes_actualizado['saldo']]
plt.scatter(clientes_actualizado['edad'], clientes_actualizado['saldo'], color=colores)
#T√≠tulo del gr√°fico y de los ejes
plt.title('Relaci√≥n entre edad y saldo de los clientes')
plt.xlabel('Edad')
plt.ylabel('Saldo')
plt.show()
````
